## Data Science I : Introduction to DB and Web Scraping

** Exercise 4 part 1: Introduction to DB and Web Scraping


Use the beautiful soup library to fetch the contents from: `www.groupce.com/python/html/thejourney.html`
- Print all \<a\> tags. How many are there?
- Print all \<tr\> tags. How many are there? (note: <tr> are tags to specify rows in an html table)
- For every \<tr\> tag, findAll \<td\> tags. Print all \<td\> tags.
- Print the second \<td\> tag found in every \<tr\> tag.

```python
import sqlite3
import urllib.request as ur
from bs4 import BeautifulSoup 
```

```python
#Open the connection to the URL
url_str = r'http://www.groupce.com/python/html/thejourney.html'
response = ur.urlopen(url_str)
```

```python
# Use the url connection to read all the data into a wariable, human readable - so use decode() function
html = response.read().decode()
```

```python
# Parsing HTML with BeautifulSoup
parsed_html = BeautifulSoup(html, 'html.parser')
#print(parsed_html)
```

```python
#1.1 Print all <a> tags. Return the count of how many are there of <a> tags.

# Extract all the 'a' tags and return a list
a_tags = parsed_html('a')
#print(a_tags)
for tag in a_tags:
    print(tag)
cnt_atag = len(a_tags)
print("Total count for <a> tag is: {}".format(cnt_atag))
```

```python
#1.2 Print all <tr> tags. Return the count of how many are there of <tr> tags.
# Extract all the 'tr' tags and return a list

tr_tags = parsed_html('tr')
cnt_trtag = 0
cnt_tdtag = 0

for tr_tag in tr_tags:
    cnt_trtag = 1 + cnt_trtag
    td_tags = tr_tag('td')
    cnt_tdtag = cnt_tdtag + len(td_tags)
    print(td_tags)
    print(td_tags[1])
    
print("Total count for <td> tag is: {}".format(cnt_tdtag))
print("Total count for <tr> tag is: {}".format(cnt_trtag))
```

## Challenge project : Web Scraping Project

** Exercise 4 part 2: Introduction to DB and Web Scraping

2.1 Create a CSV file from books information found in: http://books.toscrape.com  
Use BeautifulSoup to collect the following information from the site : Book Title, Book_Price, Book_Rating and Category  
Store all the data in a CSV file called ‘bookprices.csv’ Note : scrape only the first page of each category.  
2.2 Load the newly created CSV file into a Pandas Dataframe  
2.3 Write the complete dataframe to a table in an sqlite3 DB (to_sql method of dataframe)  
2.4 Open and read the table from Python. Execute a query to extract the MIN, MAX and AVG Price for each of the categories.


```python
#Import libraries
import urllib.request as ur
import sqlite3
from bs4 import BeautifulSoup
import csv
import pandas as pd
import sqlite3 as sqldb
```

```python
headers = {'User-Agent' : 'Mozilla/5.0 (Linux; Android 11; SAMSUNG SM-G973U) AppleWebKit/537.36 (KHTML, like Gecko) SamsungBrowser/14.2 Chrome/87.0.4280.141 Mobile Safari/537.36'}
reg_url = 'http://books.toscrape.com'
```

```python
req =  ur.Request(url=reg_url, headers=headers)
```

```python
html = ur.urlopen(req).read().decode()
```

```python
# Parsing HTML with BeautifulSoup
parsed_html = BeautifulSoup(html, 'html.parser')

# Find the categories division
category_div = parsed_html('div','side_categories')[0]
#parsed_html
#category_div
```

```python
# Extract all the 'a' tags and return a list of tuple with category name and page links 
a_category = category_div('a')
# print(a_category)
# List containing category names and url links
category_list = [] 

for tag in a_category[1:]:
    #print(tag)
    ctg_name = tag.contents[0].strip()
    ctg_link = (tag.get('href'))
    ctg_tuple = (ctg_name, ctg_link)
    category_list.append(ctg_tuple)
print(category_list[0])    
cnt_atag = len(category_list)
print("Total count for <a> tag is: {}".format(cnt_atag))
```

```python
# Get the book information that is in the form of "http://books.toscrape.com/catalogue/category/books/travel_2/index.html"
# Retrieve each category book page
book_records = []

for category in category_list:
    #print(category)
    category_name = category[0]
    category_url = reg_url + '/' + category[1]
    #print(category_name, category_url)

    #Open the connection to the URL 
    req = ur.Request(url=category_url, headers=headers)
    response = ur.urlopen(req)
    # Use the url connection to read all the data into a wariable, human readable - so use decode() function
    html = response.read().decode()
    # Parsing HTML with BeautifulSoup
    parsed_html = BeautifulSoup(html, 'html.parser')
    #print(parsed_html)
    
    # Get the 'Article' Tag, it contains inside all relevant book information
    book_info = parsed_html('article','product_pod')
    #print(book_info)
    
    for book in book_info:
        #print(book)
        book_title  = book.find("h3").text.strip()
        book_price  = float(book.find('div','product_price').find('p','price_color').contents[0].strip('£'))
        book_rating = book.find('p').get('class')[1]
        book_category =  category_name
        #print(book_title)
        #print(book_rating,book_price)
        book_tuple = (book_title, book_price , book_rating, book_category)
        book_records.append(book_tuple)
        print(book_tuple)
```

```python
# Store all the data into a csv file
# open the csv file in the write mode
f = open('sourcedata\\bookprices.csv', 'w', newline='', encoding='utf-8')

# csv writer
csvwriter = csv.writer(f)

# write a row to the csv file
csvwriter.writerow(['Book_title', 'Book_price', 'Book_rating','Book_category'])
csvwriter.writerows(book_records)

# close the file
f.close()
```

```python
#Load the newly created CSV file into a Pandas Dataframe
df_books = pd.read_csv('sourcedata\\bookprices.csv')
df_books
```

```python
#Write the complete dataframe to a table in an sqlite3 DB
cn = sqldb.connect('db\\booksdb.db')
df_books.to_sql('books', cn,  if_exists='replace');

# Close the connection 
cn.close()
```

```python
#Open and read the table from Python. Execute a query to extract the MIN, MAX and AVG Price for each of the categories.
import os

# SQL statement to select records and perform summary by categories 
sql_text = "select book_category, min(book_price) as min_price, max(book_price) as max_price, avg(book_price) as avg_price from books group by book_category"
filenm = 'booksdb.db'


dir_path = os.path.dirname(os.path.realpath('db\\booksdb.db'))
file_path = dir_path + '\\' + filenm 
print(file_path)

# connection object
cn = sqldb.connect(file_path)

# cursor object 
cursor_obj = cn.cursor() 
  
cursor_obj.execute(sql_text) 


records = cursor_obj.fetchall()
for row in records: 
    category, min_price, max_price, avg_price = row 
    print('Category = {:30}  :  minimum price = {:.2f}, maximum price = {:.2f} , average book price = {:.2f}'.format(category, min_price, max_price, avg_price))  

  
cn.commit() 

# Close the connection 
cn.close()
```

```python
# QC results with the dataframe
qc_df_result = df_books.groupby('Book_category').agg({'Book_price': ['min','max','mean']})
print("Min, max, and mean of book_price by Book_category") 
print(qc_df_result) 
```
